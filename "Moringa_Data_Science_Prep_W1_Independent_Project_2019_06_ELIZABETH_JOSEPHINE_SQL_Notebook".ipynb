```
#loading and extension into the environment
%load_ext sql
# connecting to the in memory sql database
%sql sqlite://
```

```
#importing the required pythin libraries for this notebook
#
import csv
import pandas as pd
```

```
# then from a local CSV file, we load the table to be used
#
with open('KARAMOJA.csv','r') as f:
    Dalberg = pd.read_csv(f, index_col=0, encoding='utf-8') 

# Then storing it in an SQL table of our in memory sqlite database
#
%sql DROP TABLE if EXISTS Dalberg;
%sql PERSIST Dalberg;
```

```
#Display a list of Sub Counties and their population and areas.
%%sql
SELECT "SUBCOUNTY_NAME", "POP", "Area" FROM Dalberg
```

```
#Sort the list of districts by total crop area (descending order). 
%%sql
SELECT * FROM Dalberg
GROUP BY DISTRICT_NAME
ORDER BY SUM(Crop_Area_Ha) DESC;
```

```
#Select only the Sub counties from the Moroto district, 
#order them alphabetically and show their production of sorghum. 
#
%%sql
SELECT * FROM Dalberg
WHERE DISTRICT_NAME = "MOROTO" 
ORDER BY S_Yield_Ha DESC;
```

```
#Compute the total Maize production per District. 
%%sql
SELECT *, SUM("M_Prod_Tot") FROM Dalberg
GROUP BY DISTRICT_NAME 
```

```
#Compute the number of Sub counties where Maize is produced and the total Maize production per District.
%%sql
SELECT *, SUM(M_Prod_Tot) FROM Dalberg
GROUP BY DISTRICT_NAME 
ORDER BY COUNT(SUBCOUNTY_NAME) DESC;
```

```
#Compute the overall Crop area in all Sub counties where population is over 20000.
%%sql
SELECT *, SUM("Crop_Area_Ha") FROM Dalberg WHERE POP > 20000;
```

```
#Sort the Maize production in descending order by Districts, only taking 
# account Sub counties where Maize area is larger than Sorghum area, and display the number of Sub counties per district matching that criteria. 
%%sql
SELECT *, COUNT(SUBCOUNTY_NAME) FROM Dalberg WHERE  "M_Area_Ha" > "S_Area_Ha"
GROUP BY DISTRICT_NAME
ORDER BY M_Prod_ToT DESC;
```
